## The Data Challenge for Grid Foundation Models

Foundation models require large, diverse datasets covering a wide range of operating conditions.

**Real grid data** is scarce:
- Proprietary and commercially sensitive
- Privacy constraints on load and generation profiles
- Rare events (faults, cascading failures) are, by definition, underrepresented
- Inconsistent formats across utilities and ISOs

**Simulation fills the gap** by generating unlimited labeled trajectories
under controlled conditions, with full observability of all internal states.

## Role of Simulation in the GridFM Ecosystem

| GridFM Component | Purpose | Simulation Role |
|-|-|-|
| Grid Data Zoo | Aggregate public datasets | Produce synthetic cases at scale |
| GridFMLab | Pretrain foundation models | Supply training trajectories |
| Downstream tasks | PF, OPF, contingency analysis | Provide ground truth for validation |

A **physics-faithful simulator** is the foundation that enables all three.

The simulator must be:
- Open-source (reproducibility)
- Scriptable (batch generation)
- Physics-transparent (not a black box)

## What Makes ANDES Different: Symbolic DAE Framework

Most simulators are **numerical black boxes**: equations are hard-coded in compiled routines.

ANDES uses a **hybrid symbolic-numeric** approach:

1. Models are defined as **symbolic SymPy expressions**
2. Jacobians are derived **automatically** via symbolic differentiation
3. Optimized numerical code is **generated** from the symbolic forms

This means the equations are **first-class objects** that can be:
- Inspected and printed
- Exported to other frameworks (PyTorch, JAX)
- Used directly as physics losses in PINNs

The same DAE system underpins **both** power flow (algebraic subset) and
time-domain simulation (full integration).

## ANDES Speaks ML Natively

### Symbolic Equations to PyTorch
- Extract any model's DAE equations as SymPy expressions
- Auto-convert to differentiable PyTorch functions
- Use as physics loss in PINNs with zero manual derivation

### Gymnasium RL Environment
- Built-in `AndesEnv` compatible with Stable Baselines3
- Observations: any model variable (voltages, frequencies, angles)
- Actions: any controllable setpoint (governor, exciter, renewable)
- Standard `reset()` / `step()` interface for RL researchers

### Data Export
- All state and algebraic variables as NumPy arrays
- Time series extraction as pandas DataFrames
- Direct compatibility with PyTorch DataLoader pipelines

- **DAE compaction**: Automatic removal of replaced static devices for faster simulation
- **Enhanced RL interface**: Multi-agent support, curriculum learning hooks
- **Improved I/O**: Native PSS/E 35+ format support for broader case compatibility
